```python
import pandas as pd
import bokeh
import numpy as np
%load_ext bokehlab
df1 = pd.read_csv('reference signal.lvm', sep='\t', skiprows=21, decimal=',')
df2 = pd.read_csv('silicon wafer signal.lvm', sep='\t', skiprows=21, decimal=',')
```

```python
df2 = df2[:-((df2.shape[0])-(df1.shape[0]))]
```

```python
fft_E_1 = np.fft.fft(df1['scan'])
fft_E_2 = np.fft.fft(df2['scan'])
```

```python
freq_1 = np.fft.fftfreq(len(ftt_E_1), df1['X_Value'][1] - df1['X_Value'][0])
```

```python
plot(np.angle(fft_E_1))
```
