```python
import pandas as pd
import bokeh
import numpy as np
%load_ext bokehlab
df1 = pd.read_csv('reference signal.lvm', sep='\t', skiprows=21, decimal=',')
df2 = pd.read_csv('silicon wafer signal.lvm', sep='\t', skiprows=21, decimal=',')
```

```python
df2 = df2[:-((df2.shape[0])-(df1.shape[0]))]
```

```python
ftt_E_1 = np.fft.fft(df1['scan'])
ftt_E_2 = np.fft.fft(df2['scan'])
```

```python
freq_1 = np.fft.fftfreq(len(ftt_E_1), df1['X_Value'][1] - df1['X_Value'][0])
```

```python
abs_1 = np.abs(ftt_E_1)
abs_2 = np.abs(ftt_E_2)
ang_1 = np.angle(ftt_E_1)
ang_2 = np.angle(ftt_E_2)
ang_H = -np.unwrap(ang_1)+np.unwrap(ang_2)
from scipy.constants import c
n = (1-(c*ang_H[1:]/(freq_1[1:]*2.08/1000*6.28)))
plot(freq_1[1:len(freq_1)//2], n[1:len(freq_1)//2], '.-', hover=True)
```
